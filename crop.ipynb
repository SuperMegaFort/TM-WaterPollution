```python
import matplotlib.pyplot as plt
from PIL import Image, ImageOps

# 1. Chemin vers ton image de test (remplace par une vraie image)
image_test_path = '/Users/cyriltelley/Desktop/MSE/Third_semester/TM-WaterPollution/data/08012026_120000_RCNX0002_Ziplo.jpg' 

# Ouvrir l'image
img = Image.open(image_test_path)
width, height = img.size

# 2. Couper l'image (Région d'intérêt : moitié inférieure)
box = (0, height // 2, width, height)
img_cropped = img.crop(box)

# 3. Appliquer le Padding pour la rendre carrée (sans déformation)
# On prend la plus grande dimension (la largeur) pour faire un carré parfait
# color=(128, 128, 128) ajoute des bandes grises (le gris est plus neutre que le noir pour l'IA)
max_dim = max(img_cropped.size)
img_padded = ImageOps.pad(img_cropped, size=(max_dim, max_dim), color=(128, 128, 128))

# 4. Redimensionnement final pour le modèle
img_final = img_padded.resize((224, 224))

# 5. Affichage
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img_cropped)
plt.title(f"1. Image Coupée ({img_cropped.size[0]}x{img_cropped.size[1]})")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_padded)
plt.title(f"2. Padding gris ({img_padded.size[0]}x{img_padded.size[1]})")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_final)
plt.title(f"3. Taille Modèle ({img_final.size[0]}x{img_final.size[1]})")
plt.axis('off')

plt.tight_layout()
plt.show()
```
